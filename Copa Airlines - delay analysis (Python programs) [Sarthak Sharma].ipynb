Analyzing only those origin-destination pairs where origin is different from destination.

**Contents:**
* Average departure delays of all OD pairs.
<br>

* Average arrival delays of all OD pairs.
<br>

* Bank analysis:
** Arrival delays.

```python
import pandas, numpy, matplotlib.pyplot

data_frame_2D = pandas.read_csv('Copa-FSU-02172021_comma_separated.csv')
print (data_frame_2D)
```

```python
print ("data_frame.shape = ", data_frame_2D.shape)
n_data_frame_rows = data_frame_2D.shape[0]
n_data_frame_columns = data_frame_2D.shape[1]
```

```python
OD_1D = ["PTYLIM"]   # First valid OD pair, as seen in Excel file.
for i in range (n_data_frame_rows):
    flag = 0
    for j in range (len (OD_1D)):
        if data_frame_2D ['OD'][i] == OD_1D [j]:
            flag = 1
            break
    if (flag == 0) and (data_frame_2D ["ORIG_CD"][i] != data_frame_2D ["DEST_CD"][i]):
        OD_1D.append (data_frame_2D['OD'][i])
        print(j, ". ", OD_1D[j])    
    
```

```python
n_OD = len (OD_1D)
print ("There are", n_OD, "origin-destination (OD) pairs, where origin is not the same as destination.")
```

# Analysis of average departure delays of all "origin - destination" pairs

```python
sum_departure_delay_minutes_1D = numpy.zeros (n_OD)
average_departure_delay_minutes_1D = numpy.zeros (n_OD)
departure_delay_count_1D = numpy.zeros (n_OD)

for i in range (n_OD):
    print ("Processing OD pair ", i)
    for j in range (n_data_frame_rows):
        if (OD_1D [i] == data_frame_2D ['OD'][j]) and (data_frame_2D ['DEP_DELAY_MINUTES'][j] != '?'):
            sum_departure_delay_minutes_1D [i] = sum_departure_delay_minutes_1D [i] + float (data_frame_2D['DEP_DELAY_MINUTES'][j])
            departure_delay_count_1D [i] = departure_delay_count_1D [i] + 1
```

```python
print ("Index \t OD pair \t Average departure delay (minutes)")
for i in range (n_OD):
    average_departure_delay_minutes_1D [i] = sum_departure_delay_minutes_1D [i] / departure_delay_count_1D [i]
    print (i, "\t", OD_1D [i], "\t", average_departure_delay_minutes_1D [i])
```

```python
matplotlib.pyplot.figure (figsize = (15, 8)) 
matplotlib.pyplot.title ("Average departure delay (minutes) vs. origin-destination pair")
matplotlib.pyplot.bar (OD_1D, average_departure_delay_minutes_1D)
matplotlib.pyplot.xlabel ("Origin-destination pair")
matplotlib.pyplot.ylabel ("Average departure delay (minutes)")
matplotlib.pyplot.savefig ("Plot - Average departure delay (minutes) vs origin-destination pair")
```

As expected, departure delays are mostly close to zero, or positive. There are small negative delays, as flights will usually not depart early (before scheduled time).

**Finding 10 largest average departure delays:**

```python
descend_sort_avg_dep_delay_minutes_1D = numpy.sort(average_departure_delay_minutes_1D)[::-1]

max_avg_dep_delay_indices_1D = []
max_avg_dep_delay_OD_1D = []

c = 0
while len (max_avg_dep_delay_OD_1D) < 10:
    if not numpy.isnan (descend_sort_avg_dep_delay_minutes_1D [c]):
        max_avg_dep_delay_indices_1D.append (numpy.where (average_departure_delay_minutes_1D == descend_sort_avg_dep_delay_minutes_1D [c]) [0][0])
        max_avg_dep_delay_OD_1D.append (OD_1D [max_avg_dep_delay_indices_1D [-1]])
    c = c + 1

print ("10 largest average departure delays:")    
print ("Rank \t\t OD pair \t Average departure delay (minutes)")
for i in range (len (max_avg_dep_delay_OD_1D)):
    print (i + 1, "\t\t", max_avg_dep_delay_OD_1D [i], "\t\t", average_departure_delay_minutes_1D [max_avg_dep_delay_indices_1D [i]])
```

**Finding 10 smallest average departure delays:**

```python
ascend_sort_avg_dep_delay_minutes_1D = numpy.sort(average_departure_delay_minutes_1D)

min_avg_dep_delay_indices_1D = []
min_avg_dep_delay_OD_1D = []

c = 0
while len (min_avg_dep_delay_OD_1D) < 10:
    if not numpy.isnan (ascend_sort_avg_dep_delay_minutes_1D [c]):
        min_avg_dep_delay_indices_1D.append (numpy.where (average_departure_delay_minutes_1D == ascend_sort_avg_dep_delay_minutes_1D [c]) [0][0])
        min_avg_dep_delay_OD_1D.append (OD_1D [min_avg_dep_delay_indices_1D [-1]])
    c = c + 1

print ("10 smallest average departure delays:")    
print ("Rank \t\t OD pair \t Average departure delay (minutes)")
for i in range (len (min_avg_dep_delay_OD_1D)):
    print (i + 1, "\t\t", min_avg_dep_delay_OD_1D [i], "\t\t", average_departure_delay_minutes_1D [min_avg_dep_delay_indices_1D [i]])
```

```python
sum_arrival_delay_minutes_1D = numpy.zeros (n_OD)
average_arrival_delay_minutes_1D = numpy.zeros (n_OD)
arrival_delay_count_1D = numpy.zeros (n_OD)

for i in range (n_OD):
    print ("Processing OD pair ", i)
    for j in range (n_data_frame_rows):
        if (OD_1D [i] == data_frame_2D ['OD'][j]) and (data_frame_2D ['ARR_DELAY_MINUTES'][j] != '?'):
            sum_arrival_delay_minutes_1D [i] = sum_arrival_delay_minutes_1D [i] + float (data_frame_2D['ARR_DELAY_MINUTES'][j])
            arrival_delay_count_1D [i] = arrival_delay_count_1D [i] + 1
```

```python
print ("No. \t OD pair \t Average arrival delay (minutes)")
for i in range (n_OD):
    average_arrival_delay_minutes_1D [i] = sum_arrival_delay_minutes_1D [i] / arrival_delay_count_1D [i]
    print (i, ". \t ", OD_1D [i], "\t", average_arrival_delay_minutes_1D [i])
```

```python
matplotlib.pyplot.figure (figsize = (15, 8)) 
matplotlib.pyplot.title ("Average arrival delay (minutes) vs origin-destination pair")
matplotlib.pyplot.bar (OD_1D, average_arrival_delay_minutes_1D)
matplotlib.pyplot.xlabel ("Origin-destination pair")
matplotlib.pyplot.ylabel ("Average arrival delay (minutes)")
matplotlib.pyplot.savefig ("Plot - Average arrival delay (minutes) vs origin-destination pair")
```

Arrival delays are spread to both positive and negative values, as flights can arrive early, on time, or late.

**Finding 10 largest average arrival delays:**

```python
descend_sort_avg_arr_delay_minutes_1D = numpy.sort(average_arrival_delay_minutes_1D)[::-1]

max_avg_arr_delay_indices_1D = []
max_avg_arr_delay_OD_1D = []

c = 0
while len (max_avg_arr_delay_OD_1D) < 10:
    if not numpy.isnan (descend_sort_avg_arr_delay_minutes_1D [c]):
        max_avg_arr_delay_indices_1D.append (numpy.where (average_arrival_delay_minutes_1D == descend_sort_avg_arr_delay_minutes_1D [c]) [0][0])
        max_avg_arr_delay_OD_1D.append (OD_1D [max_avg_arr_delay_indices_1D [-1]])
    c = c + 1

print ("10 largest average arrival delays:")    
print ("Rank \t\t OD pair \t average arrival delay (minutes)")
for i in range (len (max_avg_arr_delay_OD_1D)):
    print (i + 1, "\t\t", max_avg_arr_delay_OD_1D [i], "\t\t", average_arrival_delay_minutes_1D [max_avg_arr_delay_indices_1D [i]])
```

**Finding 10 smallest average arrival delays:**

```python
ascend_sort_avg_arr_delay_minutes_1D = numpy.sort (average_arrival_delay_minutes_1D)

min_avg_arr_delay_indices_1D = []
min_avg_arr_delay_OD_1D = []

c = 0
while len (min_avg_arr_delay_OD_1D) < 10:
    if not numpy.isnan (ascend_sort_avg_arr_delay_minutes_1D [c]):
        min_avg_arr_delay_indices_1D.append (numpy.where (average_arrival_delay_minutes_1D == ascend_sort_avg_arr_delay_minutes_1D [c]) [0][0])
        min_avg_arr_delay_OD_1D.append (OD_1D [min_avg_arr_delay_indices_1D [-1]])
    c = c + 1

print ("10 smallest average arrival delays:")    
print ("Rank \t\t OD pair \t average arrival delay (minutes)")
for i in range (len (min_avg_arr_delay_OD_1D)):
    print (i + 1, "\t\t", min_avg_arr_delay_OD_1D [i], "\t\t", average_arrival_delay_minutes_1D [min_avg_arr_delay_indices_1D [i]])
```

For a flight, arrival delays are probably more important than departure delays, as customer (i.e. passenger) satisfaction will depend on whether the flight arrives on time at the destination airport. Small departure delays may be taken care of by increasing aircraft's average flight speed. But arrival delays can not be undone.

# Bank analysis

```python
banks_1D = ["Off Bank"]

for i in range (n_data_frame_rows):
    flag = 0
    for j in range (len (banks_1D)):
        if data_frame_2D ['BANK OVERALL'][i] == banks_1D [j]:
            flag = 1
            break
    if flag == 0:
        banks_1D.append (data_frame_2D ['BANK OVERALL'][i])
```

```python
n_banks = len (banks_1D)
print ("There are", n_banks, "banks: \n")

for i in range (n_banks):
    print(i, ". \t", banks_1D[i])
```

## Banks: arrival delay analysis

For arrival analysis of PTY banks, the destination must be PTY.

```python
banks_sum_arrival_delay_minutes_1D = numpy.zeros (n_banks)
banks_average_arrival_delay_minutes_1D = numpy.zeros (n_banks)
banks_arrival_delay_count_1D = numpy.zeros (n_banks)

for i in range (n_banks):
    print ("Processing bank ", banks_1D [i])
    for j in range (n_data_frame_rows):
        if (banks_1D [i] == data_frame_2D ['BANK OVERALL'][j]) and (data_frame_2D ['ARR_DELAY_MINUTES'][j] != '?') and (data_frame_2D ['DEST_CD'][j] == "PTY"):
            banks_sum_arrival_delay_minutes_1D [i] = banks_sum_arrival_delay_minutes_1D [i] + float (data_frame_2D['ARR_DELAY_MINUTES'][j])
            banks_arrival_delay_count_1D [i] = banks_arrival_delay_count_1D [i] + 1
```

```python
print ("No. \t Bank \t Average arrival delay (minutes)")
for i in range (n_banks):
    banks_average_arrival_delay_minutes_1D [i] = banks_sum_arrival_delay_minutes_1D [i] / banks_arrival_delay_count_1D [i]
    print (i, ". \t ", banks_1D [i], "\t", banks_average_arrival_delay_minutes_1D [i])
```

```python
matplotlib.pyplot.figure (figsize = (15, 8)) 
matplotlib.pyplot.title ("Average arrival delay (minutes) vs. banks")
matplotlib.pyplot.bar (banks_1D, banks_average_arrival_delay_minutes_1D)
matplotlib.pyplot.xlabel ("Bank")
matplotlib.pyplot.ylabel ("Bank average arrival delay (minutes)")
matplotlib.pyplot.savefig ("Plot - Average arrival delay (minutes) vs banks")
```

```python
print ("No. \t Bank \t Average departure delay (minutes)")
for i in range (n_banks):
    banks_average_departure_delay_minutes_1D [i] = banks_sum_departure_delay_minutes_1D [i] / banks_departure_delay_count_1D [i]
    print (i, ". \t ", banks_1D [i], "\t", banks_average_departure_delay_minutes_1D [i])
```
